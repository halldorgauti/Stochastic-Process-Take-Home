<h4>Question 2</h4>
Calculate the probability of the system having a health level of 4 after 10 days.

Lets calculate the transition probability matrix $P_{(i,j)}$

```python
import numpy as np
import math
from scipy.special import factorial
import pandas as pd

def ber(k,n,p):
    return factorial(n)/factorial(k)/factorial(n-k)*p**k*(1-p)**(n-k)


mu = 4

def Pupper(i,j,p,acc=50):
    k = np.array(list(range(acc)))
    summ = np.sum(np.exp(-mu)*mu**(k+j-i)/factorial(k+j-i)*ber(k,k+j,p))
    return summ


def Plower(i,j,p,acc=50):
    k = np.array(list(range(acc)))
    summ = np.sum(np.exp(-mu)*mu**(k)/factorial(k)*ber(k+i-j,k+i,p))
    #print(np.exp(-mu)*mu**(k)/factorial(k)*ber(k+i-j,k+i,p))
    return summ

a = 10
MarkovMatr = np.zeros([a+1,a+1])


#Upper triangle including diagonal
for i in range(a):
    for j in range(i,a):
        MarkovMatr[i,j] = Pupper(i,j,0.5)

#Lower triangle not including diagonal
for i in range(1,a):
    for j in range(i):
        MarkovMatr[i,j] = Plower(i,j,0.5)

#Fill in last row of matrix
su=[1-sum(i) for i in MarkovMatr]
MarkovMatr[:,-1] = su

P = pd.DataFrame(MarkovMatr)
#Print transistion prob matrix
P
```

10 step transition matrix

```python
P_10step = pd.DataFrame(np.linalg.matrix_power(P,10))
P_10step
```

```python
print('10 step transition probability: ', P_10step[4][0])
```

Create modified transition matrix such that when reaching state a we return to state 0 in next step

```python
P_mod = P.copy()
P_mod[10][10] = 0
P_mod[0][10] = 1
#Print the modified transition matrix
P_mod
```
