```python
import numpy as np
import pandas as pd
N = 200
h = 9 #9 repair units
S = np.zeros([N,N])
lambd = 4
mu = 0.5
S[0,0:2] = [-lambd, lambd]
for i in range(1,N-1):
    if i<h:
        S[i,i-1] = mu*i
        S[i,i] = -mu*i-lambd
        S[i,i+1] = lambd
    if i>=h:
        S[i,i-1] = mu*h
        S[i,i] = -mu*h-lambd
        S[i,i+1] = lambd
S[-1,-1] = -h*mu
S[-1,-2] = h*mu
pd.DataFrame(S)
```

```python
from scipy.linalg import expm
limiting = expm(S*10000)
limiting[0,0:100]
```
